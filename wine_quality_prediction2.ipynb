## binarization of quality
- --> qulaity > 6 is 1 (good)
- --> quality <= 6 is 0 (bad)

```python
df['best quality'] = [1 if x > 5 else 0 for x in df.quality]
```

```python
# splitting data into training and tst model
features = df.drop(['quality', 'best quality'], axis=1)
target = df['best quality']
# traning the data
xtrain, xtest, ytrain, ytest = train_test_split(features, target, test_size=0.2, random_state=40)
xtrain.shape, xtest.shape
```

```python
#normalizating data before fitting into model
norm = MinMaxScaler()
xtrain = norm.fit_transform(xtrain)
xtest = norm.transform(xtest)
```

```python
models = [LogisticRegression(), XGBClassifier(), SVC(kernel='rbf'),RandomForestClassifier()]

for i in range(4):
    models[i].fit(xtrain, ytrain)

    print(f'{models[i]} : ')
    print('Training Accuracy : ', metrics.roc_auc_score(ytrain, models[i].predict(xtrain)))
    print('Validation Accuracy : ', metrics.roc_auc_score(
        ytest, models[i].predict(xtest)))
    print()
```

```python
#classification report
print(metrics.classification_report(ytest, models[2].predict(xtest)))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Assuming 'models[2]' is our trained classifier
cm = confusion_matrix(ytest, models[2].predict(xtest))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=models[2].classes_) # Assuming your model has a 'classes_' attribute
disp.plot()
plt.show()
```

```python
# visulazing the above accuracy rates
import matplotlib.pyplot as plt

# Model names and corresponding accuracies
models = ['Logistic Regression', 'XGBClassifier', 'SVC','RandomForestClassifier']
train_accuracies = [0.7595, 1.0, 0.7773,1.0]
val_accuracies = [0.7486, 0.7678, 0.7751,0.7971]

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 5))

# Bar width
bar_width = 0.35

# X-axis positions for the bars
index = range(len(models))

# Create bars for training and validation accuracies
bar1 = ax.bar(index, train_accuracies, bar_width, label='Training Accuracy', color='green')
bar2 = ax.bar([i + bar_width for i in index], val_accuracies, bar_width, label='Validation Accuracy', color='orange')

# Adding labels and title
ax.set_xlabel('Model')
ax.set_ylabel('Accuracy')
ax.set_title('Model Performance Comparison')
ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(models)
ax.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
